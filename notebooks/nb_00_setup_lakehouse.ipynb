```python
from notebookutils.fs import *

#Create Bronze File directory

mkdirs("Files/Bronze/")

#Create Script folder for config file
mkdirs("Files/Bronze/Scripts")

#Create audit schema for audit/metadata tables

mkdirs("Tables/Audit/")
mkdirs("Tables/Silver/")
mkdirs("Tables/Gold/")
```

```python
%%sql

-- Create ingestion batches table

CREATE TABLE IF NOT EXISTS audit.ingestion_batches (
    batch_id STRING,
    pipeline_name STRING,
    pipeline_run_id STRING,
    start_ts TIMESTAMP,
    end_ts TIMESTAMP,
    status STRING,
    triggered_by STRING
)
USING DELTA;
```

```python
%%sql

-- Create ingestion files table

CREATE TABLE IF NOT EXISTS audit.ingestion_files (
    batch_id STRING,
    entity_name STRING,
    source_url STRING,
    target_path STRING,
    row_count BIGINT,
    ingestion_ts TIMESTAMP,
    status STRING,
    error_message STRING
)
USING DELTA;
```

```python
%%sql

-- Create Silver.Sales table

CREATE TABLE IF NOT EXISTS silver.sales (
    SalesOrderNumber STRING NOT NULL,
    SalesOrderLineNumber INT NOT NULL,
    OrderDate DATE,
    StockDate DATE,
    CustomerID INT NOT NULL,
    ProductID INT NOT NULL,
    TerritoryID INT,
    Quantity INT,
    CreatedTS TIMESTAMP,
    ModifiedTS TIMESTAMP
)
USING DELTA;
```

```python
%%sql

-- Create Products Delta Table

CREATE TABLE IF NOT EXISTS silver.products (
    ProductKey INT NOT NULL,
    ProductSKU STRING,
    ProductName STRING,
    ModelName STRING,
    ProductDescription STRING,
    CategoryKey INT,
    CategoryName STRING,
    SubcategoryKey INT,
    SubcategoryName STRING,
    CreatedTS TIMESTAMP,
    ModifiedTS TIMESTAMP
)
USING DELTA;
```

```python
%%sql

-- Create Customers Delta Table

CREATE TABLE IF NOT EXISTS silver.customers (
    CustomerKey INT,
    Prefix STRING,
    FirstName STRING,
    LastName STRING,
    FullName STRING,
    Gender STRING,
    MaritalStatus STRING,
    BirthDate DATE,
    Email STRING,
    AnnualIncome INT,
    TotalChildren INT,
    EducationLevel STRING,
    Occupation STRING,
    HomeOwner STRING,
    CreatedTS TIMESTAMP,
    ModifiedTS TIMESTAMP
)
USING DELTA;
```

```python
%%sql

-- Create Territories Delta Table

CREATE TABLE IF NOT EXISTS silver.territories (
    TerritoryKey INT,
    Continent STRING,
    Country STRING,
    Region STRING,
    CreatedTS TIMESTAMP,
    ModifiedTS TIMESTAMP
)
USING DELTA;
```

```python
%%sql

-- Create Silver Calendar Delta Table

CREATE TABLE IF NOT EXISTS silver.calendar (
    Date DATE,
    Year INT,
    Quarter INT,
    Month INT,
    WeekOfYear INT,
    Day INT,
    DayOfWeek INT,
    DayOfYear INT,
    IsWeekend BOOLEAN,
    CreatedTS TIMESTAMP,
    ModifiedTS TIMESTAMP
)
USING DELTA;
```

```python
%%sql

-- Create Silver Returns Delta Table

CREATE TABLE IF NOT EXISTS silver.returns (
    ReturnDate DATE,
    TerritoryKey INT,
    ProductKey INT,
    ReturnQuantity INT,
    CreatedTS TIMESTAMP,
    ModifiedTS TIMESTAMP
)
USING DELTA;
```

```python
%%sql

-- Create Gold dim_date Delta Table

CREATE TABLE gold.dim_date (
    DateKey INT,
    Date DATE,
    Day INT,
    DayName STRING,
    DayOfWeek INT,
    WeekOfYear INT,
    Month INT,
    MonthName STRING,
    Quarter INT,
    Year INT,
    IsWeekend BOOLEAN
)
USING DELTA;
```

```python
%%sql

-- Create Gold dim_product Delta Table

CREATE TABLE gold.dim_product (
    ProductKey INT,
    ProductSKU STRING,
    ProductName STRING,
    ModelName STRING,
    CategoryName STRING,
    SubcategoryName STRING,
    ListPrice DECIMAL(10,2),
    StandardCost DECIMAL(10,2)
)
USING DELTA;
```

```python
%%sql

-- Create Gold dim_customer Delta Table

CREATE TABLE gold.dim_customer (
    CustomerKey INT,
    FullName STRING,
    FirstName STRING,
    LastName STRING,
    Gender STRING,
    MaritalStatus STRING,
    Age INT,
    AnnualIncome INT,
    EducationLevel STRING,
    Occupation STRING,
    HomeOwner STRING,
    Email STRING,
    EmailDomain STRING
)
USING DELTA;
```

```python
%%sql

-- Create Gold dim_territory Delta Table

CREATE TABLE gold.dim_territory (
    TerritoryKey INT,
    Region STRING,
    Country STRING,
    Continent STRING
)
USING DELTA;
```

```python
%%sql

-- Create Gold fact_sales Delta Table

CREATE TABLE gold.fact_sales (
    SalesOrderNumber STRING,
    SalesOrderLineNumber INT,
    OrderDateKey INT,
    CustomerKey INT,
    ProductKey INT,
    TerritoryKey INT,
    Quantity INT,
    UnitPrice DECIMAL(10,2),
    SalesAmount DECIMAL(12,2)
)
USING DELTA;
```

```python
%%sql

-- Create Gold fact_sales Delta Table

CREATE TABLE IF NOT EXISTS gold.fact_returns (
    ReturnDateKey INT,
    ProductKey INT,
    TerritoryKey INT,
    ReturnQuantity INT,
    UnitPrice DECIMAL(10,2),
    ReturnAmount DECIMAL(12,2)
)
USING DELTA;
```
