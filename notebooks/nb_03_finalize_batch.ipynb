```python
# Set a parameter that will pull pipeline variable value - current batch_id

batch_id=""
```

```python
from notebookutils import runtime
from datetime import datetime

# Parameter injected: batch_id

end_ts = datetime.utcnow()

# Determine batch status
failed_count = spark.sql(f"""
SELECT COUNT(*) AS cnt 
FROM audit.ingestion_files
WHERE batch_id = '{batch_id}' AND status = 'FAILED'
""").collect()[0]['cnt']

batch_status = "FAILED" if failed_count > 0 else "SUCCESS"

# Update batch table
spark.sql(f"""
UPDATE audit.ingestion_batches
SET end_ts = TIMESTAMP('{end_ts}'),
    status = '{batch_status}'
WHERE batch_id = '{batch_id}'
""")
```
